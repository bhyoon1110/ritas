# FTIR Spectrum Viewer
DPT 파일(wavenumber, intensity 2열 CSV)을 읽어 Plotly 인터랙티브 그래프로 시각화합니다.
- 마우스 스크롤: 줌인/줌아웃
- 드래그: 범위 이동
- 더블클릭: 원래 보기로 복귀
- 툴바 카메라 아이콘: SVG 이미지 저장

```python
import pandas as pd
import plotly.graph_objects as go
```

## 1. 파일 경로 설정
`FILE_PATH` 에 분석할 `.dpt` 파일 경로를 입력하세요.

```python
# 분석할 DPT 파일 경로 (변경 가능)
FILE_PATH = "sample_TR(transmittance).0.dpt"

# 샘플 이름 (그래프 제목 및 범례에 표시)
SAMPLE_NAME = "sample_TR"
```

## 2. 데이터 로드 및 미리보기

```python
df = pd.read_csv(FILE_PATH, header=None, names=["wavenumber", "intensity"])
df = df.sort_values("wavenumber").reset_index(drop=True)

print(f"포인트 수: {len(df):,}")
print(f"Wavenumber 범위: {df['wavenumber'].min():.1f} ~ {df['wavenumber'].max():.1f} cm⁻¹")
print(f"Intensity 범위: {df['intensity'].min():.5f} ~ {df['intensity'].max():.5f}")
df.head()
```

## 3. 인터랙티브 스펙트럼 그래프

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df["wavenumber"],
    y=df["intensity"],
    mode="lines",
    name=SAMPLE_NAME,
    line=dict(color="#2563eb", width=1.5),
    hovertemplate="<b>%{x:.1f} cm⁻¹</b><br>Intensity: %{y:.5f}<extra></extra>",
))

fig.update_layout(
    title=dict(
        text=f"FTIR Spectrum — {FILE_PATH}",
        font=dict(size=18),
        x=0.01,
    ),
    xaxis=dict(
        title="Wavenumber (cm⁻¹)",
        range=[df["wavenumber"].max(), df["wavenumber"].min()],  # 역방향 (FTIR 관례)
        showgrid=True,
        gridcolor="#e8e8e8",
        tickmode="linear",
        dtick=500,
        minor=dict(showgrid=True, gridcolor="#f4f4f4"),
    ),
    yaxis=dict(
        title="Intensity",
        showgrid=True,
        gridcolor="#e8e8e8",
    ),
    plot_bgcolor="white",
    paper_bgcolor="#fafafa",
    height=550,
    hovermode="x",
    font=dict(family="Arial, sans-serif"),
    margin=dict(l=70, r=30, t=60, b=60),
)

fig.show()
```

## 4. HTML 파일로 저장 (선택)

```python
out_html = FILE_PATH.replace(".dpt", "_plot.html")

fig.write_html(
    out_html,
    include_plotlyjs="cdn",
    config={
        "scrollZoom": True,
        "toImageButtonOptions": {
            "format": "svg",
            "filename": SAMPLE_NAME,
            "scale": 2,
        },
    },
)

print(f"저장 완료: {out_html}")
```
